```python
xx, yy = np.meshgrid(np.linspace(0, 1, len(stat)) * ome_max, np.linspace(0, 1, len(stat)) * phi_max)
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xx, yy, np.rad2deg(np.amax(stat, axis=0)),
                cmap=cm.coolwarm,
                linewidth=0, 
                antialiased=False)
ax.set_xlabel('$\omega$')
ax.set_ylabel('$\phi$')
ax.set_title('Responnse magnitude')
```
