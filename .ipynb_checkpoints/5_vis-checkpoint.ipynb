```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
%matplotlib notebook

from swing_config import *
from swing_plot import swing_plot
from swing_open_loop import fv_gen
```

```python
stat = np.load('./data/stat.npy')
```

```python
xx, yy = np.meshgrid(np.linspace(0, 1, len(stat)), np.linspace(0, 1, len(stat)))
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xx, yy, np.rad2deg(np.amax(stat, axis=2)))
ax.set_xlabel('$A$')
ax.set_ylabel('$\omega$')
ax.set_title('Responnse magnitude')
```

```python
xx, yy = np.meshgrid(np.linspace(0, 1, len(stat)), np.linspace(0, 1, len(stat)))
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xx, yy, np.rad2deg(np.amax(stat, axis=1)))
ax.set_xlabel('$\phi$')
ax.set_ylabel('$A$')
ax.set_title('Responnse magnitude')
plt.show()
```
